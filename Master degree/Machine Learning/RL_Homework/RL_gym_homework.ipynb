### Create environment and some random actions

```python
import gym_space
env = gym.make('MySpaceInvaders-v0')
env.reset()
print('Initial environment:')
env.render()
for episode in range(1,3):
  print('\nEpisode: ', episode)
  for _ in range(15):
      env.step(env.action_space.sample()) # take a random action
      env.render()
  env.reset()
env.close()
```
