```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

img1 = cv2.imread('../12.1.png', 0)
img2 = cv2.imread('../12.2.png', 0)

def computeGradients(img): 
    # Kernels
    sobelx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    sobely = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    scharrx = np.array([[3,0,-3],[10,0,-10],[3,0,-3]])
    scharry = np.array([[3,10,3],[0,0,0],[-3,-10,-3]])
    prewittx = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
    prewitty = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    robertsx = np.array([[1,0],[0,-1]])
    robertsy = np.array([[0,1],[-1,0]])
    
    # Compute Sobel: derivatives and gradient
    sox = convolve2d(img, sobelx)
    soy = convolve2d(img, sobely)
    soG = np.sqrt(sox**2 + soy**2)

    # Compute Scharr: derivatives and gradient
    scx = convolve2d(img, scharrx)
    scy = convolve2d(img, scharry)
    scG = np.sqrt(scx**2 + scy**2)

    # Compute Prewitt: derivatives and gradient
    prx = convolve2d(img, prewittx)
    pry = convolve2d(img, prewitty)
    prG = np.sqrt(prx**2 + pry**2)

    # Compute Roberts: derivatives and gradient
    rox = convolve2d(img, robertsx)
    roy = convolve2d(img, robertsy)
    roG = np.sqrt(rox**2 +roy**2)
    
    return soG, scG, prG, roG

soG1, scG1, prG1, roG1 = computeGradients(img1)
soG2, scG2, prG2, roG2 = computeGradients(img2)

# Save and show (in next cells)
cv2.imwrite("soG1.png", soG1)
cv2.imwrite("scG1.png", scG1)
cv2.imwrite("prG1.png", prG1)
cv2.imwrite("roG1.png", roG1)

cv2.imwrite("soG2.png", soG2)
cv2.imwrite("scG2.png", scG2)
cv2.imwrite("prG2.png", prG2)
cv2.imwrite("roG2.png", roG2)
```
